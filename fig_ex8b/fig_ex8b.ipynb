```python
import scanpy as sc
import os
import sys
import numpy as np
sys.path.append(os.pardir)

model_name = 'normal'
results_analysis_file = '../data/calculated/preprocessing/{}_analysis.h5ad'.format(model_name)
seed = 0

# # if you need
import warnings
warnings.filterwarnings('ignore')
```

```python
adata_analysis = sc.read(results_analysis_file)
```

```python
adata_analysis.raw = sc.AnnData(
    np.log2(adata_analysis.raw.X.toarray() + 1),
    var=adata_analysis.raw.var,
    obs=adata_analysis.obs
)
```

```python
# sepcific RNA
target_gene_list = ['Krt14','Krt1', 'Lor', 'Mki67', 'Krt79', 
                    'Cd34','Krt75','Gli1','Mgst1','Lgr6',
                    'Dct', 'Kit','Cldn10','Cd3d','Cd207']

for g in target_gene_list:
    print(g)
    sc.pl.umap(adata_analysis, color=[g], components = ['1,2'], ncols=1, size=2, cmap='inferno')
```
